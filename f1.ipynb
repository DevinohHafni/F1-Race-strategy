```python
import fastf1
import pandas as pd

fastf1.Cache.enable_cache('cache')

races = ['Monaco', 'Silverstone', 'Spa', 'Bahrain']
year = 2023

all_laps = []

for race in races:
    try:
        print(f"Loading {race}...")
        session = fastf1.get_session(year, race, 'R')
        session.load()
        
        laps = session.laps
        laps['Race'] = race
        
        all_laps.append(laps)
        
    except Exception as e:
        print(f"Error loading {race}: {e}")

df = pd.concat(all_laps, ignore_index=True)

print(df.head())
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
df['Sector1Time'] = df['Sector1Time'].astype('timedelta64[ns]')
df['Sector1SessionTime'] = df['Sector1SessionTime'].astype('timedelta64[ns]')

df = df.dropna(subset=['Sector1Time', 'Sector1SessionTime'])
```

```python
df.isnull().sum()
```

```python
df['LapTime_sec'] = df['LapTime'].dt.total_seconds()
df = df[(df['LapTime_sec'] > 50) & (df['LapTime_sec'] < 150)]
df['TireAge'] = df['TyreLife']
df['StintID'] = df['Driver'] + "_" + df['Stint'].astype(str) + "_" + df['Race']
df = df.sort_values(by=['Driver', 'Race', 'LapNumber'])
df['LapInStint'] = df.groupby('StintID').cumcount() + 1
df = df[df['LapTime_sec'] < df['LapTime_sec'].quantile(0.95)]
```

```python
df_clean = df[
    (df['LapTime_sec'] > 60) & 
    (df['LapTime_sec'] < 120)
]
deg_data = df_clean.groupby(['Compound', 'LapInStint'])['LapTime_sec'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

for compound in deg_data['Compound'].unique():
    subset = deg_data[deg_data['Compound'] == compound]
    
    plt.plot(subset['LapInStint'], subset['LapTime_sec'], label=compound)

plt.xlabel('Lap in Stint')
plt.ylabel('Lap Time (seconds)')
plt.title('Tire Degradation Analysis')
plt.legend()
plt.show()
```
